▶ 결혼정보업체 데이터셋

- **gender**: 성별(female, male)

- **age**: 나이 (my age)

- **age_partner**: 상대방 나이  (partner's age)

- **importance_same_religion**: 종교를 중요시 여기는 지에 대한 응답 (1-10)

- **pref_of_partner_attractive**: 매력(attractive)에 대하여 상대방이 얼마나 중요하게 생각하는지의 여부

- **pref_of_partner_sincere**: 진정성/성실성(sincere)에 대하여 상대방이 얼마나 중요하게 생각하는지의 여부

- **pref_of_partner_funny**: 유머감각(funny)에 대하여 상대방이 얼마나 중요하게 생각하는지의 여부

- **pref_of_partner_shared_interests**: 공통의 관심사(shared interests)에 대하여 상대방이 얼마나 중요하게 생각하는지의 여부

- **eval_by_partner_attractive**: 상대방이 나의 매력(attractive)에 대하여 평가한 값

- **eval_by_partner_sincere**: 상대방이 나의 진정성/성실성(sincere)에 대하여 평가한 값

- **eval_by_partner_funny**: 상대방이 나의 유머감각(funny)에 대하여 평가한 값

- **eval_by_partner_shared_interests**: 상대방이 나와 얼마만큼의 공통 관심사(shared interests)가 있는지를 평가한 값

- **my_pref_attractive**: 매력(attractive)에 대하여 내가 얼마나 중요하게 생각하는지의 여부

- **my_pref_sincere**: 진정성/성실성(sincere)에 대하여 내가 얼마나 중요하게 생각하는지의 여부

- **my_pref_funny**: 유머감각에 대하여 내가 얼마나 중요하게 생각하는지의 여부

- **my_pref_shared_interests**: 공통 관심사에 대하여 내가 얼마나 중요하게 생각하는지의 여부

- **my_eval_attractive**: 내가 상대방의 매력에 대하여 평가한 값

- **my_eval_sincere**: 내가 상대방의 진정성/성실성에 대하여 평가한 값

- **my_eval_funny**: 내가 상대방의 유머감각에 대하여 평가한 값

- **my_eval_shared_interests**: 내가 상대방과의 얼마만큼의 공통 관심사가 있는지를 평가한 값

- **interests_correlate**: 서로간의 공통 관심사 연관도

- **how_much_i_liked**: 내가 상대방이 얼마나 마음에 들었는지의 여부(값)

- **guess_prob_liked**: 상대방이 나를 얼마나 마음에 들어하였을지의 여부(예상값)

- **married**: 만남 이후 결혼 성사 여부
