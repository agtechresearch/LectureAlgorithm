## 3. 전처리: (1) 결측치 처리

```python
# 항목별 결측치 비율 확인
married_dataset.isna().mean()
```

```python
# 결측치 처리
married_dataset = married_dataset.dropna(axis=0)
married_dataset
```

```python
# object 타입의 문자열 변수를 숫자형으로 변환
married_dataset = pd.get_dummies(married_dataset, columns=['gender'], drop_first=True)
```

```python
married_dataset.head(5)
```

## 4. 모델 학습 및 평가


```python
# 데이터셋, 독립변수와 종속변수 분리: 독립변수 -> x, 종속변수 -> y
x = married_dataset.drop(['married'], axis=1)
y = married_dataset['married']
```

```python
x
```

```python
y
```

```python
# train set 과 test set 으로 데이터를 나누기 위해 train_test_split 함수를 불러옴
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=100)
```

```python
import xgboost as xgb

#model = xgb.XGBClassifier(n_estimators=500, max_depth=5, random_state=100)
model = xgb.XGBClassifier(random_state=100)
model.fit(X_train, y_train)
```

```python
pred = model.predict(X_test)  # 테스트 데이터로 예측
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
accuracy_score(y_test, pred)  # 정확도
```

```python
print(confusion_matrix(y_test, pred))  # confusion matrix 출력
```

- 실제값이 0 인데 0 으로 예측한 경우가 935 건으로 가장 많음

- 결혼에 성공한 경우를 제대로 예측한 경우도 83 건으로 무난한 수준임

- 1종 오류, 즉 실제값은 0 인데 1 로 예측한 경우가 63 건

- 2종 오류, 즉 실제값은 1 인데 0 으로 예측한 경우가 130 건

```python
print(classification_report(y_test, pred))  # classification report 출력
```

## 5. 하이퍼파라미터 튜닝 (Hyper-parameter tuning): Grid search 

```python
from sklearn.model_selection import GridSearchCV
```

```python
# 하이퍼파라미터 튜닝을 위한 후보 파라미터들의 범위(set) 설정
parameters = {
    'learning_rate': [0.1, 0.2, 0.3],  # 학습률
    'max_depth': [2, 3, 5],             # 각 트리의 최대 깊이 제한
    'subsample': [0.5, 0.7, 0.8],         # 각 트리마다 사용할 훈련 데이터의 비율 (예 0.5를 쓰면 데이터의 절반씩만 랜덤 추출하여 트리 생성) -> 오버피팅 방지에 사용
    'n_estimators': [100, 300, 500]    # 전체 트리의 개수
}
```

```python
# CPU 코어 8개를 사용하여 학습
model_tuning = xgb.XGBClassifier(random_state=100)
gridsearch_model = GridSearchCV(model_tuning, parameters, scoring='f1', cv=5, n_jobs=8, verbose=1)
# scoring='f1' -> f1 score 로 평가
# cv=5 -> 5-fold cross validation
# verbose=1 -> 학습 과정 출력

## GPU 사용 시
#model_tuning = xgb.XGBClassifier(tree_method='gpu_hist', gpu_id=0)
#gridsearch_model = GridSearchCV(model_tuning, parameters, scoring='f1', cv=5, verbose=1)
```

```python
gridsearch_model.fit(X_train, y_train)
```

* 학습 완료까지 꽤 많은 시간이 소요됨
* 하이퍼파라미터 4종류 당 3개씩: 총 81번 모델링 진행(3^4) + 교차검증 5회 진행 => 총 405번 모델링 작업 수행

```python
gridsearch_model.best_params_
```

```python
pred_tuning = gridsearch_model.predict(X_test)  # 테스트 데이터로 예측
accuracy_score(y_test, pred_tuning)  # 정확도
```
