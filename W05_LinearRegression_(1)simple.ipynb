# 선형회귀 모델 실습: Linear Regression simple example

```python
import numpy as np
import matplotlib.pyplot as plt
import warnings

# 경고 메시지 출력 표기 생략
warnings.filterwarnings('ignore')
```

## 샘플에 활용할 데이터 셋 만들기

```python
def make_linear(w=3, b=5, size=50, noise=1.0):
    x = 4 * np.random.rand(size)
    y = w * x + b
    noise = np.random.uniform(-abs(noise), abs(noise), size=y.shape)
    yy = y + noise
    plt.figure(figsize=(10, 7))
    #plt.plot(x, y, color='r', label=f'y = {w}x + {b}')
    plt.scatter(x, yy, label='data')
    #plt.legend(fontsize=15)
    plt.show()
    print(f'w: {w}, b: {b}')
    return x, yy
```

```python
x, y = make_linear(w=2.0, b=5.0, size=100, noise=3.0)
```

```python
x = x.reshape(100,1)
x
```

```python
y
```

## Linear regression 모델 훈련

## Scikit-learn 라이브러리의 LinearRegression() 사용

```python
from sklearn.linear_model import LinearRegression

# 선형회귀 모델
model = LinearRegression()

# 모델 훈련
model.fit(x, y)
```

```python
# 회귀계수(w)
model.coef_
```

```python
# 편향(b) 혹은 y 절편
model.intercept_
```

```python
# 예측
y_pred = model.predict(x) 

# 선형 회귀선 그리기
plt.scatter(x, y)
plt.plot(x, y_pred)
```
